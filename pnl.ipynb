```python
import pandas as pd

# Load the csv file
file_path = '/content/drive/MyDrive/Solana/export_transfer_3vQvwGiGcwVTQMgaqnaKNyzNyPcQVwH39cW3J7yYtZ1w_1720453850095.csv'  # Update with your file path
df = pd.read_csv(file_path)
```

```python
df
```

```python
# Initial bet size and initial index value
initial_bet_size = 2000000
initial_index_value = 100000000

# Calculate PNL excluding initial bet size
df['pnl'] = df['Amount'] - initial_bet_size

# Exclude rows where pnl is zero
df_filtered = df[df['pnl'] != 0]

# Calculate returns based on initial index value
df_filtered['Return'] = df_filtered['pnl'] / initial_bet_size

# Calculate cumulative returns
df_filtered['cumulative_return'] = df_filtered['Return'].cumsum()
```

```python
# Calculate cumulative returns (equity curve)
df_filtered['equity_curve'] = 1 + (1 * df_filtered['cumulative_return'])
```

```python
import matplotlib.pyplot as plt

# Assuming df_filtered and its required columns are already defined

# Adjust equity curve to start from the first value as 1
first_value = df_filtered['equity_curve'].iloc[0]
df_adjusted = df_filtered['equity_curve'] / first_value

# Plot adjusted equity curve
plt.figure(figsize=(12, 8))
plt.plot(df_filtered.index, df_adjusted, label='Equity Curve', color='b')

plt.xlabel('Trade Index')
plt.ylabel('Value (Normalized)')
plt.title('Performance Analysis')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Calculate daily returns
df_filtered['daily_return'] = df_filtered['equity_curve'].pct_change()

# Plot daily returns
plt.figure(figsize=(12, 6))
plt.plot(df_filtered.index, df_filtered['daily_return'], label='Daily Return', color='m')
plt.xlabel('Trade Index')
plt.ylabel('Daily Return')
plt.title('Daily Returns Over Time')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Calculate drawdown
df_filtered['previous_peak'] = df_filtered['equity_curve'].cummax()
df_filtered['drawdown'] = (df_filtered['equity_curve'] - df_filtered['previous_peak']) / df_filtered['previous_peak']

# Plot drawdown
plt.figure(figsize=(12, 6))
plt.plot(df_filtered.index, df_filtered['drawdown'], label='Drawdown', color='r')
plt.xlabel('Trade Index')
plt.ylabel('Drawdown')
plt.title('Drawdown Over Time')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming df_filtered and its required columns are already defined

# Calculate rolling statistics
rolling_mean = df_filtered['daily_return'].rolling(window=20).mean()


# Plot rolling statistics
plt.figure(figsize=(12, 8))
plt.plot(df_filtered.index, df_filtered['daily_return'], label='Daily Return', color='m', alpha=0.5)
plt.plot(df_filtered.index, rolling_mean, label='30-Day Rolling Mean', color='g')

# Highlight zero daily return line
plt.axhline(y=0, color='b', linestyle='-', linewidth=2, label='Zero Daily Return')

plt.xlabel('Trade Index')
plt.ylabel('Daily Return')
plt.title('Daily Returns and Rolling Statistics')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np

# Assuming df_filtered and its required columns are already defined

# Calculate daily returns
df_filtered['daily_return'] = df_filtered['equity_curve'].pct_change()

# Risk-free rate assumption (you should set this to the actual risk-free rate applicable to your trading period)
risk_free_rate = 0.02  # For example, assuming 2% annual risk-free rate

# Calculate Sharpe Ratio
daily_returns = df_filtered['daily_return']
sharpe_ratio = np.sqrt(252) * (daily_returns.mean() - risk_free_rate) / daily_returns.std()

# Calculate Sortino Ratio
downside_returns = daily_returns[daily_returns < 0]
sortino_ratio = np.sqrt(252) * (daily_returns.mean() - risk_free_rate) / downside_returns.std()

# Calculate Calmar Ratio
max_drawdown = df_filtered['drawdown'].min()
calmar_ratio = daily_returns.mean() / abs(max_drawdown)

# Print metrics
print(f'Sharpe Ratio: {sharpe_ratio:.2f}')
print(f'Sortino Ratio: {sortino_ratio:.2f}')
print(f'Calmar Ratio: {calmar_ratio:.2f}')
```
